```python
import pandas as pd
import numpy as np
import math
```

```python

table = pd.read_excel('Data.xlsx')
table.columns
cleanTable = table.drop(['Unnamed: 0'], axis=1)
cleanTable
```

```python
cleanTable.columns
```

```python
cols = ['Order_Number','Carrier','Service_Class','Service_Description','Package_Weight','Package_Length','Package_Width','Package_Height','Zone']
x = cleanTable[cols]
x
```

```python

#df.round({'A': 1, 'C': 2})
x = x.round({'Package_Length':0, 'Package_Width':0,'Package_Height':0})
x['Rounded_Package_Weight'] = x['Package_Weight'] + .5
```

```python
x
x = x.round({'Rounded_Package_Weight':0})
x
```

```python
x['Dimensional_Weight']= (x['Package_Length']*x['Package_Width']*x['Package_Height']/139) + .5
x
```

```python
x = x.round({'Dimensional_Weight':0})
x
```

```python

for index, row in x.iterrows():
    if row['Rounded_Package_Weight'] > row['Dimensional_Weight']:
        row['Chargeable_Weight'] = ['Rounded_Package_Weight']
    else:
        row['Chargeable_Weight'] = ['Dimensional_Weight']

x
```
